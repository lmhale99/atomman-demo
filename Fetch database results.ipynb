```python
import iprPy

import matplotlib.pyplot as plt
```

```python
# Load settings for a database
database = iprPy.load_database('master')
```

```python
# Retrieve all finished E_vs_r_scan results for fcc Ag
records = database.get_records('calculation_E_vs_r_scan', family='A1--Cu--fcc', composition='Ag', status='finished')
print(len(records), 'matching records found')
```

```python
fig = plt.figure(figsize=(14,6))

for record in records[:10]:
    plt.plot(record.r_values, record.energy_values, label=record.potential.potential_LAMMPS_id)
    
plt.xlim(2, 6)
plt.ylim(-4, 2)

plt.xlabel('Interatomic spacing (angstrom)', size='x-large')
plt.ylabel('Potential energy (eV/atom)', size='x-large')

plt.legend(fontsize='small')

plt.savefig('images/escan.png')
```
